# Generation of noisy images

## Import

```python
RUN_DIR = 'tf/'
```

```python
import os
import shutil
import sys
import time

from skimage import img_as_ubyte
from skimage.io import imsave, imread_collection
from skimage.util import random_noise
```

## Global Variables

```python
NO_NOISY_DATASET_PATH: str = RUN_DIR + 'data_ae/no_noisy'
SAVE_TRAIN_PATH: str = RUN_DIR + 'data_ae/noisy/training/photo'
SAVE_TEST_PATH: str = RUN_DIR + 'data_ae/noisy/validation/photo'
```

## Remove and/or create directories for noisy training and validation dataset

```python
def remove_and_create_dir(path):
    if os.path.exists(path):
        shutil.rmtree(path)
    os.makedirs(path)

remove_and_create_dir(SAVE_TRAIN_PATH)
remove_and_create_dir(SAVE_TEST_PATH)
```

## Generate noisy dataset and save it

```python
def progressbar(it, prefix="", size=60, file=sys.stdout):
    count = len(it)

    def show(j):
        x = int(size * j / count)
        file.write("%s[%s%s] %i/%i\r" % (prefix, "#" * x, "." * (size - x), j, count))
        file.flush()
        file.write("\n")

    show(0)
    for i, item in enumerate(it):
        yield i, item
        show(i + 1)
    file.flush()


def noisy_gauss_and_save(images, path, name):
    for i, image in progressbar(images, f"Add noise and save image {name} : ", 50):
        time.sleep(0.1)
        image = random_noise(image, mode='gaussian', mean=0, var=0.3)
        image = random_noise(image, mode='s&p', amount=0.2, salt_vs_pepper=0.5)
        image = random_noise(image, mode='poisson')
        image = random_noise(image, mode='speckle', mean=0, var=0.1)
        imsave(os.path.join(path, f"photo_{i + 1:04d}.jpg"), img_as_ubyte(image))


def generate_noisy_dataset(path, save_path, name):
    images = imread_collection(path + "/*.jpg")
    images = noisy_gauss_and_save(images, save_path, name)
    return images


# Train noisy
train_data_noisy = generate_noisy_dataset(NO_NOISY_DATASET_PATH + "/training/photo", SAVE_TRAIN_PATH, "training")

# Test noisy
test_data_noisy = generate_noisy_dataset(NO_NOISY_DATASET_PATH + "/validation/photo", SAVE_TEST_PATH, "validation")
```
